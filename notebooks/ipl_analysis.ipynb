# 4. Data Cleaning


```python
# Fix the data types
matches['date'] = pd.to_datetime(matches['date'])
matches['season'] = matches['date'].dt.year

# Handle the missing values
# 'winner' is NaN for no result matches - Filling with "No Result"
matches['winner'] = matches['winner'].fillna('No Result')

# 'player_of_match' can be unknown 
matches['player_of_match'] = matches['player_of_match'].fillna("Unknown")

# 'city' - We can fill the venue too but some of the venue don't have city name so i will fill it with 'Unknown'
matches['city'] = matches['city'].fillna('Unknown')

# I just observed something that the name of the teams have changed too like 'Delhi Daredevils' to 'Delhi Capitals'
# So I would fix that too

# -------- Standarizing team names ( franchise renamed over the years ) ----
team_name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Kings XI Punjab': 'Punjab Kings'
}

for col in ['team1','team2','winner','toss_winner']:
    matches[col] = matches[col].replace(team_name_map)

for col in ['batting_team','bowling_team']:
    deliveries[col] = deliveries[col].replace(team_name_map)

print("Data Cleaning done")
print(f"Seasons covered: {sorted(matches['season'].unique())}")
print(f"Total matches: {len(matches)}")
print(f"Total deliveries: {len(deliveries)}")
```

## 5. Match Level Analysis

```python
# ------ Matches per season ----
matches_per_season = matches.groupby('season').size().reset_index(name="matches")

plt.figure(figsize=(12,5))
sns.barplot(data=matches_per_season, x='season',y='matches',hue='season',palette='Blues_d',legend=False )
plt.title('Number of matches per IPL season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------- Wins by runs VS Wins by wickets -----------
win_type = matches[matches['winner'] != 'No Result']['result'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(win_type,labels=win_type.index, autopct='%1.1f%%',
        colors=['#4C72B0', '#DD8452'],
        startangle=90,
        wedgeprops={'edgecolor': 'white', 'linewidth': 2})
plt.title("Matches Result: Win by runs VS wickets",fontweight='bold')
plt.tight_layout()
plt.show()
```

## 6. Team Performance analysis


```python
# ----- Total wins per team - Top 10 (excludes no results) --------
team_wins = (matches[matches['winner'] != 'No Result']['winner']
            .value_counts()
            .reset_index()
            .rename(columns={'winner':'team','count':'wins'})
            .head(10))

plt.figure(figsize=(12,6))
sns.barplot(data=team_wins,x='wins',y='team',hue ='team',palette='viridis')
plt.title("Top 10 teams by total winnings (all seasons)",fontsize=16,fontweight='bold')
plt.xlabel("Total Wins")
plt.ylabel("Teams")
plt.tight_layout()
plt.show()
```

```python
# ----- Win rate per team --------
total_matches = pd.concat([matches['team1'],matches['team1']]).value_counts().reset_index()
total_matches.columns = ['team','total']

wins = matches[matches['winner'] != 'No Result']['winner'].value_counts().reset_index()
wins.columns = ['team','wins']

win_rate = total_matches.merge(wins, on='team')
win_rate['win_rate'] = (win_rate['wins'] / win_rate['total'] * 100).round(2)

# Getting only the win_rate where total matches are above the 20 
win_rate = win_rate[win_rate['total'] >= 20].sort_values('win_rate', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=win_rate, x='win_rate', y='team' ,hue='team', palette = 'magma')
plt.title("Win Rate of the Teams (min 20 matches)", fontsize=16, fontweight='bold')
plt.xlabel('Win Rate (%)')
plt.ylabel('Team')
plt.axvline(50, color = 'red', linestyle='--',linewidth=2,label='50% Mark')
plt.legend()
plt.show()
```

# 4. Data Cleaning


```python
# Fix the data types
matches['date'] = pd.to_datetime(matches['date'])
matches['season'] = matches['date'].dt.year

# Handle the missing values
# 'winner' is NaN for no result matches - Filling with "No Result"
matches['winner'] = matches['winner'].fillna('No Result')

# 'player_of_match' can be unknown 
matches['player_of_match'] = matches['player_of_match'].fillna("Unknown")

# 'city' - We can fill the venue too but some of the venue don't have city name so i will fill it with 'Unknown'
matches['city'] = matches['city'].fillna('Unknown')

# I just observed something that the name of the teams have changed too like 'Delhi Daredevils' to 'Delhi Capitals'
# So I would fix that too

# -------- Standarizing team names ( franchise renamed over the years ) ----
team_name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Kings XI Punjab': 'Punjab Kings'
}

for col in ['team1','team2','winner','toss_winner']:
    matches[col] = matches[col].replace(team_name_map)

for col in ['batting_team','bowling_team']:
    deliveries[col] = deliveries[col].replace(team_name_map)

print("Data Cleaning done")
print(f"Seasons covered: {sorted(matches['season'].unique())}")
print(f"Total matches: {len(matches)}")
print(f"Total deliveries: {len(deliveries)}")
```

## 5. Match Level Analysis

```python
# ------ Matches per season ----
matches_per_season = matches.groupby('season').size().reset_index(name="matches")

plt.figure(figsize=(12,5))
sns.barplot(data=matches_per_season, x='season',y='matches',hue='season',palette='Blues_d',legend=False )
plt.title('Number of matches per IPL season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# -------- Wins by runs VS Wins by wickets -----------
win_type = matches[matches['winner'] != 'No Result']['result'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(win_type,labels=win_type.index, autopct='%1.1f%%',
        colors=['#4C72B0', '#DD8452'],
        startangle=90,
        wedgeprops={'edgecolor': 'white', 'linewidth': 2})
plt.title("Matches Result: Win by runs VS wickets",fontweight='bold')
plt.tight_layout()
plt.show()
```

## 6. Team Performance analysis


```python
# ----- Total wins per team - Top 10 (excludes no results) --------
team_wins = (matches[matches['winner'] != 'No Result']['winner']
            .value_counts()
            .reset_index()
            .rename(columns={'winner':'team','count':'wins'})
            .head(10))

plt.figure(figsize=(12,6))
sns.barplot(data=team_wins,x='wins',y='team',hue ='team',palette='viridis')
plt.title("Top 10 teams by total winnings (all seasons)",fontsize=16,fontweight='bold')
plt.xlabel("Total Wins")
plt.ylabel("Teams")
plt.tight_layout()
plt.show()
```

```python
# ----- Win rate per team --------
total_matches = pd.concat([matches['team1'],matches['team1']]).value_counts().reset_index()
total_matches.columns = ['team','total']

wins = matches[matches['winner'] != 'No Result']['winner'].value_counts().reset_index()
wins.columns = ['team','wins']

win_rate = total_matches.merge(wins, on='team')
win_rate['win_rate'] = (win_rate['wins'] / win_rate['total'] * 100).round(2)

# Getting only the win_rate where total matches are above the 20 
win_rate = win_rate[win_rate['total'] >= 20].sort_values('win_rate', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=win_rate, x='win_rate', y='team' ,hue='team', palette = 'magma')
plt.title("Win Rate of the Teams (min 20 matches)", fontsize=16, fontweight='bold')
plt.xlabel('Win Rate (%)')
plt.ylabel('Team')
plt.axvline(50, color = 'red', linestyle='--',linewidth=2,label='50% Mark')
plt.legend()
plt.show()
```

## 7. Batting Analysis

```python
# --------- Top 10 runs scorer -----------
top_batters = (deliveries.groupby('batter')['batsman_runs']
              .sum()
              .sort_values(ascending=False)
              .head(10)
              .reset_index()
              .rename(columns={'batsman_runs':'total_runs'}))

plt.figure(figsize=(12,6))
sns.barplot(data=top_batters, x='total_runs', y='batter',palette='husl',hue='batter')
plt.title('Top 10 run scorers in IPL history',fontsize=16,fontweight='bold')
for i in range(len(top_batters)):
     plt.text(top_batters['total_runs'][i] + 50, i,
             str(top_batters['total_runs'][i]),
             va='center')
plt.xlabel('Total Runs')
plt.ylabel('Batter')
plt.tight_layout()
plt.show()
```

```python
# ------- Boundary Analysis ----------
fours = deliveries[deliveries['batsman_runs'] == 4].groupby('batter').size()
sixes = deliveries[deliveries['batsman_runs'] == 6].groupby('batter').size()

boundary_df = pd.DataFrame({'fours': fours, 'sixes': sixes}).fillna(0).astype(int)
boundary_df['Total boundaries'] = boundary_df['fours'] + boundary_df['sixes']
top_boundary = boundary_df.sort_values('Total boundaries',ascending=False).head(10).reset_index()

fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(top_boundary))
width = 0.4
ax.bar(x - width/2, top_boundary['fours'], width, label='Fours', color='#4C72B0')
ax.bar(x + width/2, top_boundary['sixes'], width, label='Sixes', color='#DD8452')
ax.set_xticks(x)
ax.set_xticklabels(top_boundary['batter'], rotation=45, ha='right')
ax.set_title('Top 10 Batters: Fours vs Sixes', fontsize=15, fontweight='bold')
ax.set_ylabel('Count')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# --------- Average runs per ball by over --------------
over_runs = (deliveries.groupby('over')['total_runs']
        .mean()
        .reset_index()
        .rename(columns={'total_runs':'avg_run'}))

plt.figure(figsize=(12,5))
sns.lineplot(data=over_runs,x='over',y='avg_run',marker ='o',color='#4C72B0',linewidth=2)
plt.fill_between(over_runs['over'],over_runs['avg_run'],alpha=0.2, color = '#4C72B0')
plt.title('Average runs per Ball by Over across all matches', fontsize=15,fontweight='bold')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.xticks(range(1,21))
plt.tight_layout()
plt.show()
```

## 8. Bowling Analysis

```python
# -------- Top 10 wicket takers ---------
# Exclude run outs as they aren't credited to the bowler
wickets = deliveries[(deliveries['is_wicket'] == 1) & (deliveries['dismissal_kind'] != 'run out')]

top_bowlers = (wickets.groupby('bowler').size()
              .sort_values(ascending=False)
              .head(10)
              .reset_index()
              .rename(columns={0:'wickets'}))
top_bowlers
plt.figure(figsize=(12,6))
sns.barplot(data=top_bowlers,x=top_bowlers.iloc[:,1],y='bowler', palette='flare',hue='bowler')
plt.title('Top 10 wicket takers in IPL history', fontsize=15,fontweight='bold')
plt.xlabel('Total Wickets')
plt.ylabel('Bowler')
plt.tight_layout()
plt.show()
```

```python
# ------- Most common dismissal types ------
dismissals = (deliveries[deliveries['is_wicket'] == 1]['dismissal_kind']
             .value_counts()
             .reset_index()
             )

plt.figure(figsize=(10,6))
sns.barplot(data=dismissals,x='count',y='dismissal_kind',palette='coolwarm',hue='dismissal_kind')
plt.title('Most Common Dismissal Types in IPL',fontsize='16',fontweight='bold')
plt.xlabel('Counts')
plt.ylabel('Dismissal Type')
plt.tight_layout()
plt.show()
```

## 9. Toss analysis

```python
# ------- Toss decision: Batting or Fielding ---------
toss_decision = matches['toss_decision'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(toss_decision,labels=toss_decision.index , autopct='%1.1f%%',
        colors=['skyblue','green'],startangle=90,
        wedgeprops = {'edgecolor':'white','linewidth':2})
plt.title('Toss Decision : Batting vs Fielding',fontsize=14,fontweight='bold')
plt.tight_layout()
plt.show()
```
